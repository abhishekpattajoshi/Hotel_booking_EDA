```python
#data visualisation
#around 37% of bookings are cancelled which is very high
cancel_perc=df['is_canceled'].value_counts(normalize=True)
cancel_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not_canceled','canceled'],df['is_canceled'].value_counts(),width=0.7)
plt.show()
```

```python
#the cancellation of reservation in city hotel is more compare to the cancellation resort hotel.This may be because the price
#of resort hotel is more compare to city hotel.We will find out the reasons in further visualisations. 
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title('Reservation status in different hotels',size=15)
plt.xlabel('hotel')
plt.ylabel('number_of_reservations')
plt.show()
```

```python
#27.9% of booking is getting cancelled in case of resort booking
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#41.7% of bookings are getting canceled in case of city hotel
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#we can clearly see that the price of the resort hotels are more compare to the city hotels. The spike in the price may be of
#weekends where the booking in resort hotel gets increased. This proofs our hypothesis testing that resort has high price compare to city.
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in resort and city hotel')
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
```

```python
#cancellation is highest in the month of january and lowest in the month of august. It may be because the rate is higher in 
#the month of january and lowest in the month of august.
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.title('Resesrvation status per month')
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.show()
```

```python
#here we can clearly see that the price of hotel in the month of january is highest due to which hotel booking is getting cancelled. 
#This proves our second hypothesis.
plt.figure(figsize=(15,8))
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR per month')
plt.show()
```

```python
#portugal is the country where majority of the bookings get cancelled. The hotels should think about giving offers and promotion
#so that the cancellation will get decrease in portugal. They should also think about reducing the prices and giving more 
#facilties at same price.
canceled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]#this will give top 10 countries
plt.figure(figsize=(8,8))
plt.title('Top 10  cancellation countries')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
```
